```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('Salary_Data.csv')
dataset = np.array([data.loc[:,'YearsExperience'],data.loc[:,'Salary']], np.float32)
print(dataset)

```

```python
years = dataset[0,:]
salary = dataset[1,:]
X,Y = years,salary
```

## Metoda Gradientului

```python
dw1 = lambda w1, w2, x, y: -sum((y[i] - (w1 * x[i] + w2)) * x[i] for i in range(len(x))) / len(x)
dw2 = lambda w1, w2, x, y: -sum(y[i] - (w1 * x[i] + w2) for i in range(len(x))) / len(x)


w = np.random.uniform(-1, 1, 2)
epochs = 1000
c = 0.01

while abs((w[0] - c * dw1(w[0], w[1], X, Y)) - w[0]) > 0.001 and abs((w[1] - c * dw2(w[0], w[1], X, Y)) - w[1]) > 0.001:      
    w[0] = w[0] - c * dw1(w[0], w[1], X, Y)
    w[1] = w[1] - c * dw2(w[0], w[1], X, Y)
    print(f"w1={w[0]}, w2={w[1]}")

plt.scatter(X, Y, label="Training data")        
plt.plot(X, w[1] + w[0] * X, color="red", label="Regression line")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Regresie liniara")
plt.show()
```
